#Data exploration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("HR.csv")
```

```python
def dataoveriew(df, message):
    print(f'{message}:\n')
    print('Number of rows: ', df.shape[0])
    print("\nNumber of features:", df.shape[1])
    print("\nData Features:")
    print(df.columns.tolist())
    print("\nMissing values:", df.isnull().sum().values.sum())
    print("\nUnique values:")
    print(df.nunique())
    
dataoveriew(df, 'Overview of the dataset')
```

```python
#deleting not used variables (emplyee count, employee number, over 18, Standard Hours)
df = df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', "StandardHours"], axis=1)
```

```python
#find out how imbalanced our target variable is
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.countplot(x='Attrition',data=df)
plt.title("Distribution of Attrition", size=14)
```

```python
#Explore outliers in numerical features
#Findings: There are no outliers in the features that have a close to normal distribution. In the non normally distibuted features we should not remove outlires I think. There are no values that seem to be wrongly recorded.
num_columns = ['Age','DailyRate', 'DistanceFromHome', 'HourlyRate', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked', 'PercentSalaryHike', 'TotalWorkingYears', 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion', 'YearsWithCurrManager']
for col in num_columns:
    plt.figure()
    sns.boxplot(x=df[col])
```

```python
#Explore the relationship with categorical variables
cat_columns = ['BusinessTravel', 'Department', 'Education', 'EducationField','EnvironmentSatisfaction', 'Gender', 'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction', 'MaritalStatus', 'OverTime', 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel', 'WorkLifeBalance']
for col in cat_columns:
    plt.figure()
    sns.countplot(x=col, hue="Attrition", data=df,)
```
