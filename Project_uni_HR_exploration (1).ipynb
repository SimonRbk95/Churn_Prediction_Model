```python
#deleting not used variables (emplyee count, employee number, over 18, Standard Hours)
df = df.drop(['EmployeeCount', 'EmployeeNumber', 'Over18', "StandardHours"], axis=1)
```

```python
#transform categorical variables using dummies 
df = pd.get_dummies(df, columns=["Attrition", "BusinessTravel", "Department", "EducationField", "Gender", "JobRole", "MaritalStatus", "OverTime"], drop_first=True)
```

```python
#find out how imbalanced our target variable is (We need to use over or under sampling if there is a big imbalance)
import seaborn as sns
sns.countplot(x='Attrition_Yes',data=df)
```
